### Writing CSV

```python
import csv
books = [
    {"title": "Python 101",   "author": "Ali",    "year": 2026},
    {"title": "Data Magic",   "author": "Nabiha", "year": 2025},
    {"title": "AI Basics",    "author": "Omar",   "year": 2024},
]
# Column names (dictionary ki keys)
headers = ["title", "author", "year"]
```

```python
print(books)
print(type(books))
```

```python
for row in books:
    print(row["title"] , "by", row["author"] , "(" , row["year"] , ")" )
```

```python
print(books[0])
print(books[1])
print(books[2])
```

```python
import csv
headers = ["title" , "author" , "year"]
with open("books.csv", "w", newline="") as f:
    writer = csv.DictWriter(f, fieldnames=headers)
    writer.writeheader()          # pehli line: headers
    for b in books:
        writer.writerow(b)        # har dict = ek r
```

### Reading CSV

```python
import csv

with open("books.csv", "r") as f:
    content = csv.DictReader(f)
    for row in content:
        # har row ek dictionary banti hai — keys = headers
        print(row["title"], "+", row["author"], "+", row["year"])
        
```
